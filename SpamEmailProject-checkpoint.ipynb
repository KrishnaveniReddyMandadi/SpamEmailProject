# Train Test Split to perform model evaluation 

```python
from sklearn.model_selection import train_test_split

X = df.drop(['spam'], axis=1)
y = df.spam
Train_X, Test_X, Train_y, Test_y = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)

print('Original set  ---> ',X.shape,y.shape,'\nTraining set  ---> ',Train_X.shape,Train_y.shape,'\nTesting set   ---> ', Test_X.shape,'', Test_y.shape)
```

Correlation plots between variables reveal a great deal about their relationship. However, due to the huge feature space, it is challenging to obtain meaningful insights. To generate fewer characteristics that cover the most variability, we can use Feature Extraction Techniques.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler

# run an LDA and use it to transform the features

lda = LinearDiscriminantAnalysis()
Train_X_lda = lda.fit(Train_X, Train_y).transform(Train_X)
Test_X_lda = lda.fit(Test_X, Test_y).transform(Test_X)

print(X.shape, y.shape)

```

```python
Evaluation_Results = pd.DataFrame(np.zeros((5,5)), columns=['Accuracy', 'Precision','Recall','F1-score','AUC-ROC score'])
Evaluation_Results.index=['Logistic Regression (LR)', 'Support Vector Machine (SVM)','Ensemble Boosting with SVM (SvmBoost)'
                          ,'Ensemble Bagging with SVM (SvmBag)', 'Gradient Boosting (GB)']
Evaluation_Results
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score, roc_auc_score, roc_curve, precision_score, recall_score
def Classification_Summary(pred,pred_prob,i):
    Evaluation_Results.iloc[i]['Accuracy']=round(accuracy_score(Test_y, pred),3)*100   
    Evaluation_Results.iloc[i]['Precision']=round(precision_score(Test_y, pred),3)*100
    Evaluation_Results.iloc[i]['Recall']=round(recall_score(Test_y, pred),3)*100
    Evaluation_Results.iloc[i]['F1-score']=round(f1_score(Test_y, pred),3)*100
    Evaluation_Results.iloc[i]['AUC-ROC score']=round(roc_auc_score(Test_y, pred),3)*100
    print('{}{}\033[1m Evaluating {} \033[0m{}{}\n'.format('<'*3,'-'*35,Evaluation_Results.index[i], '-'*35,'>'*3))
    print('Accuracy = {}%'.format(round(accuracy_score(Test_y, pred),3)*100))
    print('F1 Score = {}%'.format(round(f1_score(Test_y, pred, average='weighted'),3)*100))
    print('\n \033[1mConfusiton Matrix:\033[0m\n',confusion_matrix(Test_y, pred))
    print('\n\033[1mClassification Report:\033[0m\n',classification_report(Test_y, pred))
    
    roc_curve(Test_y, pred_prob)
    plt.show()
def AUC_ROC_plot(Test_y, pred):    
    ref = [0 for _ in range(len(Test_y))]
    ref_auc = roc_auc_score(Test_y, ref)
    lr_auc = roc_auc_score(Test_y, pred)

    ns_fpr, ns_tpr, _ = roc_curve(Test_y, ref)
    lr_fpr, lr_tpr, _ = roc_curve(Test_y, pred)

    plt.plot(ns_fpr, ns_tpr, linestyle='--')
    plt.plot(lr_fpr, lr_tpr, marker='.', label='AUC = {}'.format(round(roc_auc_score(Test_y, pred)*100,2))) 
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.show()

```

```python

from sklearn.linear_model import LogisticRegression

LR = LogisticRegression().fit(Train_X_lda, Train_y)
pred = LR.predict(Test_X_lda)
pred_prob = LR.predict_proba(Test_X_lda)
pred_prob = pred_prob[:, 1]
Classification_Summary(pred,pred_prob,0)
AUC_ROC_plot(Test_y, pred)

```

```python
from sklearn.svm import SVC

SVM = SVC(probability=True).fit(Train_X_lda, Train_y)
pred = SVM.predict(Test_X_lda)
pred_prob = SVM.predict_proba(Test_X_lda)
pred_prob = pred_prob[:, 1]
Classification_Summary(pred,pred_prob,1)
AUC_ROC_plot(Test_y, pred)
```

```python
from sklearn.ensemble import AdaBoostClassifier
SvmBoost = AdaBoostClassifier(base_estimator=SVM, n_estimators=50, random_state=314).fit(Train_X_lda, Train_y)
pred = SvmBoost.predict(Test_X_lda)
pred_prob = SvmBoost.predict_proba(Test_X_lda)
pred_prob = pred_prob[:, 1]
Classification_Summary(pred,pred_prob,2)
AUC_ROC_plot(Test_y, pred)
```

```python
from sklearn.ensemble import BaggingClassifier
SvmBag = BaggingClassifier(base_estimator=SVM, n_estimators=31, random_state=314).fit(Train_X_lda, Train_y)
pred = SvmBag.predict(Test_X_lda)
pred_prob = SvmBag.predict_proba(Test_X_lda)
pred_prob = pred_prob[:, 1]
Classification_Summary(pred,pred_prob,3)
AUC_ROC_plot(Test_y, pred)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GB = GradientBoostingClassifier().fit(Train_X_lda, Train_y)
pred = GB.predict(Test_X_lda)
pred_prob = GB.predict_proba(Test_X_lda)
pred_prob = pred_prob[:, 1]
Classification_Summary(pred,pred_prob,4)
AUC_ROC_plot(Test_y, pred)
```

```python
#Plotting Confusion-Matrix of all the predictive Models

labels=['Ham','Spam']

def plot_cm(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred, labels=np.unique(y_true))
    cm_sum = np.sum(cm, axis=1, keepdims=True)
    cm_perc = cm / cm_sum.astype(float) * 100
    annot = np.empty_like(cm).astype(str)
    nrows, ncols = cm.shape
    for i in range(nrows):
        for j in range(ncols):
            c = cm[i, j]
            p = cm_perc[i, j]
            if i == j:
                s = cm_sum[i]
                annot[i, j] = '%.1f%%\n%d/%d' % (p, c, s)
            elif c == 0:
                annot[i, j] = ''
            else:
                annot[i, j] = '%.1f%%\n%d' % (p, c)
    cm = pd.DataFrame(cm, index=np.unique(y_true), columns=np.unique(y_true))
    cm.columns=labels
    cm.index=labels
    cm.index.name = 'Actual'
    cm.columns.name = 'Predicted'
    #fig, ax = plt.subplots()
    sns.heatmap(cm, annot=annot, fmt='')
    
def conf_mat_plot(all_models):
    plt.figure(figsize=[20,7])
    
    for i in range(len(all_models)):
        plt.subplot(2,4,i+1)
        pred = all_models[i].predict(Test_X_lda)
        plot_cm(Test_y, pred)
        plt.title(Evaluation_Results.index[i])
    plt.tight_layout()
    plt.show()

conf_mat_plot([LR,SVM,SvmBoost,SvmBag,GB])
```
